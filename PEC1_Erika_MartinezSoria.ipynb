### Respuesta:

La encuesta de Kaggle denominada 2021 Kaggle Machine Learning & Data Science Survey (Kaggle, 2021a)  abarca temas como el perfil del científico de datos, su educación, la experiencia en aprendizaje automático y ciencia de datos y situación laboral, obteniendo respuestas de alrededor de 25 000 data scientists alrededor del mundo.

Nuestro objetivo con este proyecto de minería de datos es el de describir los factores determinantes en el ingreso anual de un data scientist a partir de diversas variables encontradas en el conjunto de datos, como, por ejemplo: género, país de residencia, nivel de educación, cargo específico, años escribiendo código, número de empleados en la compañía en que trabaja, número de empleados en el equipo de analítica, herramientas que utilizan diariamente, etc. La descripción lo hacemos con el fin de encontrar asociaciones significativas o causales entre las diferentes variables. 

Después de haber establecido el objetivo analítico, nos encargamos de la preparación de los datos. En un primer paso, Kaggle ya ha realizado parte de la limpieza del conjunto de datos. Según el documento de la metodología de la encuesta, para asegurar la calidad de los datos, se excluyó a las encuestas que el sistema detectó como “Spam” o “Duplicado”, también se eliminó las respuestas de los encuestados que tardaron menos de dos minutos, así como las que seleccionaron menos de 15 opciones de respuesta (Kaggle, 2021b). De esta manera, ya encargándose de los datos incompletos y redundantes, parte del proceso de limpieza faltante sería el encargarse de revisar los datos incorrectos o inconsistentes y errores de transcripción. La otra sub-fase de preparación de datos es la transformación, la cual tiene el fin de poner los datos en una forma adecuada para la futura aplicación del modelo. En el proyecto podría ser que necesitemos transformar las variables categóricas y numéricas a valores binarios (Rasinen, citado en Andrew, 2012) en el caso de que usemos redes bayesianas. En el caso de la reducción de dimensionalidad, el cual utilizamos para reducir el número de atributos y registros del conjunto de datos para mejorar la eficiencia del algoritmo de minería de datos, el proyecto necesitaría pasar de tener 368 atributos a un número que refleje los más relevantes para su utilización en el modelo descriptivo. Por ejemplo, podría determinarse que la variable con el tipo de hardware especializado que utiliza un científico de datos tenga menos importancia que la variable con el país de origen, por lo tanto, se lo puede descartar. 

Finalizada la fase de preparación de los datos, la cual deja los datos con la calidad adecuada, entramos en la fase de construcción de modelos. Esta fase la conceptualizamos como un proceso de búsqueda del mejor modelo que debe estar basada en los datos que tenemos y el conocimiento a priori (Minguillón, 2016). Por ejemplo, sabemos que países desarrollados ofrecen un mayor salario que países en desarrollo, por lo que asumimos que la variable de ingreso debe tener asociación significativa con el país de residencia. Utilizando los modelos disponibles para nuestro objetivo analítico, los que pueden ser redes bayesianas o las reglas de asociación, escogemos el modelo que exprese mejor esta relación. En la sub-fase de evaluación e interpretación del modelo, donde medimos la calidad del modelo, utilizaríamos la mitad de los 25000 registros para construir el modelo y el resto para evaluarlo. Esto lo realizamos con el objetivo de medir hasta que punto se ajustan al dominio descrito (Minguillón, 2016).

Vemos el numero de muertos dependiendo del tipo de intersección en el que ocurrió el accidente.

```python
#Seleccionamos las 5 tipos de intersecciones donde han sucedido mas accidentes
intersecciones = accidentData['TYP_INTNAME'][ (accidentData['TYP_INTNAME'] == "Not an Intersection") | (accidentData['TYP_INTNAME'] == "Four-Way Intersection") | (accidentData['TYP_INTNAME'] == "T-Intersection") | (accidentData['TYP_INTNAME'] == "Y-Intersection") | (accidentData['TYP_INTNAME'] == "Reported as Unknown")]

#Creamos un crosstable con las variables de las muertes y de los tipos de intersecciones
cross_fatals_intersection = pd.crosstab(accidentData["FATALS"], intersecciones) 

#Creamos el steacked bar graph
ax = cross_fatals_intersection.plot.bar(stacked=True) 
ax.set_title("Relacion entre número de muertos y tipo de intersección", size = 16)
ax.set_xlabel("Numero de muertes", size = 16)
ax.set_ylabel("Cuenta", size = 16)
```

En esta gráfica podemos ver que la mayoría de muertes no se producen en una intersección, por lo que podemos concluir que al menos este tipo de infrasestructura en carreteras no influye en la mayoría de accidentes. Sin embargo, la interseccion de 4 vías le sigue en el número de accidentes ocurridos.

Vemos el numero de muertos dependiendo de proximidad el cruce en el que ocurrió el accidente.

```python
#Seleccionamos las 5 tipos de cruces donde han sucedido mas accidentes
cruces = accidentData['RELJCT2NAME'][ (accidentData['RELJCT2NAME'] == "Non-Junction") | (accidentData['RELJCT2NAME'] == "Intersection") | (accidentData['RELJCT2NAME'] == "Intersection-Related") | (accidentData['RELJCT2NAME'] == "Driveway Access Related") | (accidentData['RELJCT2NAME'] == "Entrance/Exit Ramp Related") | (accidentData['RELJCT2NAME'] == "Driveway Access Related") | (accidentData['RELJCT2NAME'] == "Entrance/Exit Ramp Related") | (accidentData['RELJCT2NAME'] == "Through Roadway")]

#Creamos un crosstable con las variables de las muertes y de los cruces
cross_fatals_cruces = pd.crosstab(accidentData["FATALS"], cruces) 

#Creamos el steacked bar graph
ax = cross_fatals_cruces.plot.bar(stacked=True) 

ax.set_title("Relacion entre número de muertos y cruce", size = 16)
ax.set_xlabel("Numero de muertes", size = 16)
ax.set_ylabel("Cuenta", size = 16)
```

De manera similar al anterior gráfico, vemos que los accidentes no occuren cerca de un cruce. Pero si es el caso, de que un numero menor de accidentes ocurren en intersecciones o relacionada con intersecciones.

Veamos los conductores bebidos y no bebidos por cada día de la semana. 

```python
#Copiamos la columna DRUNK_DR en una nueva columna CATEGORIA
accidentData["CATEGORIA"] = accidentData["DRUNK_DR"]

#Reemplazamos los valores 1,2,3,4 con la categoria Bebido
accidentData_drink = accidentData["CATEGORIA"].replace([1, 2, 3, 4], "Bebido")

#Reemplazamos el valor 0 con la categoria No bebido
accidentData_drink = accidentData_drink.replace(0, "No bebido")

#Crosstab de las variables con los dias de la semana y con los conductores bebidos o no bebidos.
crosstab = pd.crosstab(accidentData_drink, accidentData['DAY_WEEKNAME'])

#Con .T los subconjuntos van en las columnas, y los conjuntos superiores estan en las filas.
df = crosstab
df = df.T  #.T : Cambia el orden, las columnas por filas y viceversa.
df = df / len(accidentData_drink) #Dividimos para ponerlos en porcentajes

#Dibujamos la crosstable
ax = df.plot.bar()

ax.set_title("Distribución por día de la semana y conductores bebidos", size = 16)
ax.set_xlabel("Días de la semana", size = 16)
ax.set_ylabel("Porcentaje", size = 16)


#accidentData.drop(columns=['CATEGORIA'], inplace=True) 
```

Podemos en un comienzo intuir que habrá más conductores bebidos los fines de semana, y en efecto, así es el caso. ### Dicretización

Discretamos las siguientes variables para la utilización en la creación de modelos.

Utilizamos la discretización con el método de partición basado en el algoritmo de k-means en la variable VE_FORMS que indica el numero de vehiculos implicados en un accidente. 

```python
accidentData['VE_FORMS'].describe()
```

Lo realizamos con k-means porque las distancias entre valores son grandes. Como vemos el valor máximo es 59 y la media es de 1.54. 

```python
#Importamos la libreria
from sklearn.preprocessing import KBinsDiscretizer

# histograma de los datos originales
plt.hist(accidentData['VE_FORMS'])
plt.show()

#Discretizamos la variable VE_FORMS que tiene el numero de vehiculos en movimiento implicados en 3 intervalos. 
VE_FORMS_DIS = KBinsDiscretizer(n_bins=3, encode='ordinal',
                           strategy = "kmeans").fit_transform(accidentData[['VE_FORMS']])

#Convertimos el array en dataframe 
VE_FORMS_DIS = pd.DataFrame(VE_FORMS_DIS)
VE_FORMS_DIS = VE_FORMS_DIS.rename(columns = {0: 'VE_FORMS_DIS'})

# Histograma de los datos transformados
VE_FORMS_DIS.hist()

#Vemos los valores en que han sido discretizados.
VE_FORMS_DIS.drop_duplicates()
```

Podemos observar que al escoger 3 intervalos, los valores son transformados en 0.0, 1.0 y 2.0. 

Ahora vamos a realizar la discretizacion de la variable PVH_INVL la cual indica el numero de vehicuos estacionados implicados, utilizando el mismo método anterior, el método de partición basado en el algoritmo de k-means.

```python
accidentData["PVH_INVL"].describe()
```

Como podemos observar sigue existiendo gran distancia entre los valores como en el caso de la variable con vehiculos en movimiento implicados, siendo en este caso el máximo valor el 11 y la media 0.04. Por lo tanto, usamos el método de partición basado en el algoritmo de k-means.

```python
# histograma de los datos originales
plt.hist(accidentData['PVH_INVL'])
plt.show()

#Discretizamos la variable VE_FORMS que tiene el numero de vehiculos en movimiento implicados en 3 intervalos. 
PVH_INVL_DIS = KBinsDiscretizer(n_bins=3, encode='ordinal',
                           strategy = "kmeans").fit_transform(accidentData[['PVH_INVL']])

#Convertimos el array en dataframe 
PVH_INVL_DIS = pd.DataFrame(PVH_INVL_DIS)
PVH_INVL_DIS = PVH_INVL_DIS.rename(columns = {0: 'PVH_INVL_DIS'})

# Histograma de los datos transformados
PVH_INVL_DIS.hist()

#Vemos los valores en que han sido discretizados.
PVH_INVL_DIS.drop_duplicates()
```

De igual manera que en el caso anterior, los valores en los que se discretiza son 0.0, 1.0 y 2.0, al haber determinado la discretización para 3 intervalos.
